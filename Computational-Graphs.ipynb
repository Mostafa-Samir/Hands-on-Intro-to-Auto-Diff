# Examples Using Computational Graphs

```python
import warnings

import numpy as np
import compgraph as cg

%matplotlib inline
warnings.filterwarnings('ignore')
```

$$f(x) = (\sin x)^{\sin x}$$

```python
cg.reset()  # resets the nodes counts to 0 to start from 0 at each visualization

x = cg.variable(np.pi / 4, name='x')

a = cg.sin(x)
b = cg.sin(x)

y = a ** b

print("y = {}".format(y))

cg.visualize_at(y, figsize=(10, 4))
```

$$f(x) = x^2 2^x$$

```python
cg.reset()

x = cg.variable(0.5, name='x')
y = (x ** 2) * (2 ** x)

print("y = {}".format(y))

cg.visualize_at(y, figsize=(10, 4))
```

$$f(x, y, z) = \sin\left(x ^{y + z}\right) - 3z\ln x^2y^3$$

```python
cg.reset()

x = cg.variable(0.5, 'x')
y = cg.variable(4, 'y')
z = cg.variable(-2.3, 'z')

f = cg.sin(x ** (y + z)) - 3 * cg.log((x ** 2) * (y ** 3))

print("f = {}".format(f))

cg.visualize_at(f, figsize=(10, 6))
```
