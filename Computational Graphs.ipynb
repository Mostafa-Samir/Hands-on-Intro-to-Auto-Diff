# Examples Using Computational Graphs

```python
import numpy as np
import compgraph as cg

%matplotlib inline
import matplotlib.pyplot as plt
```

$$f(x) = (\sin x)^{\sin x}$$

```python
x = cg.variable(np.pi / 4, name='x')

a = cg.sin(x)
b = cg.sin(x)

y = a ** b

print "y = %s" % (y)

plt.rcParams['figure.figsize'] = (10, 4)
cg.visualize_at(y)
```

$$f(x) = x^2 2^x$$

```python
x = cg.variable(0.5, name='x')
y = (x ** 2) * (2 ** x)

print "y = %s" % y

plt.rcParams['figure.figsize'] = (10, 4)

cg.visualize_at(y)
```

$$f(x, y, z) = \sin\left(x ^{y + z}\right) - 3z\ln x^2y^3$$

```python
x = cg.variable(0.5, 'x')
y = cg.variable(4, 'y')
z = cg.variable(-2.3, 'z')

f = cg.sin(x ** (y + z)) - 3 * cg.log((x ** 2) * (y ** 3))

print "f = %s" % (f)

plt.rcParams['figure.figsize'] = (10, 6)
cg.visualize_at(f)
```
